```python
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
import torch

model_id1="dreamlike-art/dreamlike-diffusion-1.0"
model_id2="stabilityai/stable-diffusion-xl-base-1.0"

pipe= StableDiffusionPipeline.from_pretrained(model_id1, torch_dtype=torch.float16,use_safetensors=True)
pipe=pipe.to("cuda")

```

*IMAGE WILL BE GENERATED WITHIN 17 sec (approx)*

Click on the run button below

```python
# Generate the image using the text-to-image pipeline
# The 'pipe' variable should already be initialized with a Stable Diffusion model
image = pipe(prompt).images[0]  # Get the first generated image from the output

# Print the prompt used for generation to the console for reference
print("[PROMPT]: ", prompt)

# Display the generated image using Matplotlib
plt.imshow(image)  # Show the image
plt.axis('off')  # Hide the axis to focus on the image
plt.show()  # Render the image in a window
```
